# Messages

LangChain 提供了一个标准消息类型，适用于所有模型提供者，确保无论调用哪种模型都能保持一致行为。

## 基本用途
使用消息的最简单方式是创建消息对象，并在调用时传递给模型。


```python
import os

from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

load_dotenv()

model = init_chat_model(
    model="qwen3.5-plus",
    model_provider="openai",
    base_url=os.getenv("DASHSCOPE_BASE_URL"),
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    temperature=0.7,
    timeout=30,
    max_tokens=1000,
)
```

```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage

system_msg = SystemMessage("You are a helpful assistant.")
human_msg = HumanMessage("Hello, how are you?")

# Use with chat models
messages = [system_msg, human_msg]
response = model.invoke(messages)
print(type(response))
print(response)
```

## 文本提示

文本提示是字符串——非常适合简单的生成任务，不需要保留对话历史。

```python
response = model.invoke("Write a haiku about spring")
```

## 消息提示 Type
可以通过提供消息对象列表，将消息列表传递给模型。

系统消息 ——告诉模型如何行为并为交互提供上下文
人类消息 ——表示用户输入和与模型的交互
AI 消息 ——模型生成的响应，包括文本内容、工具调用和元数据
工具消息 ——表示工具调用的输出

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("You are a poetry expert"),
    HumanMessage("Write a haiku about spring"),
    AIMessage("Cherry blossoms bloom...")
]
response = model.invoke(messages)
print(type(response))
print(response)
```

### System message  系统消息
SystemMessage 代表一组初始指令，用于启动模型的行为。
你可以用系统信息来定调，定义模型的角色，并制定回应的指导方针。

```python
system_msg = SystemMessage("You are a helpful coding assistant.")

messages = [
    system_msg,
    HumanMessage("How do I create a REST API?")
]
response = model.invoke(messages)
print(response.pretty_print())
```

### Human message  人类信息
HumanMessage 代表用户的输入和交互。它们可以包含文本、图片、音频、文件以及其他任何数量的多模态内容 。

Message metadata  消息元数据，本质是给对话消息附加额外标识信息
* name作用：标识消息的发送者（用户 / 角色）
* id作用：给每条消息分配唯一 ID，用于全链路追踪


```python

alice_msg = HumanMessage(content="帮我查订单", name="alice")
bob_msg = HumanMessage(content="帮我退款", name="bob")

response =model.invoke([alice_msg, bob_msg])
print(response.pretty_print())
```

### AI message  AI 信息 
AIMessage 可以在其 usage_metadata 字段中存储代币计数和其他使用元数据

```python
ai_msg = AIMessage("I'd be happy to help you with that question!")

messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,  # Insert as if it came from the model
    HumanMessage("Great! What's 2+2?")
]

response = model.invoke(messages)
response.usage_metadata
```

### Streaming and chunks  流式传输与区块


```python
chunks = []
full_message = None
for chunk in model.stream("Hi"):
    chunks.append(chunk)
    print(chunk.content, end="", flush=True)
    full_message = chunk if full_message is None else full_message + chunk
```

### Tool message  工具信息
工具消息用于将单个工具执行的结果反馈给模型。


```python
from langchain.messages import ToolMessage


ai_message = AIMessage(
    content=[],
    tool_calls = [
        {
            "name":"get_weather",
            "args": {"location": "San Francisco"},
            "id": "call_123",
        }
    ]
)

weather_result = "Sunny, 72°F"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="call_123"  # Must match the call ID
)

messages = [
    HumanMessage("What's the weather in San Francisco?"),
    ai_message,  # Model's tool call
    tool_message,  # Tool execution result
]

response = model.invoke(messages)
print(response.pretty_print())
```

## Message content  信息内容
多模态内容，图片
把消息的内容看作是发送到模型的数据有效载荷，
消息有一个内容属性， 该属性是松散类型的，支持字符串和未类型对象列表（例如词典），
使得 LangChain 聊天模型中可以直接支持提供者原生结构，如多模态内容和其他数据。

```python
from langchain.messages import HumanMessage

human_message = HumanMessage("Hello, how are you?")

human_message = HumanMessage(content=[
    {"type": "text", "text": "Hello, how are you? Do you know who she are in this photo"},
    {"type": "image_url", "image_url": {"url": "https://c-ssl.dtstatic.com/uploads/blog/202207/31/20220731170533_48bad.thumb.400_0.jpeg"}}
])

human_message = HumanMessage(content_blocks=[
    {"type": "text", "text": "Hello, how are you? Do you know who she are in this photo"},
    {"type": "image", "url": "https://c-ssl.dtstatic.com/uploads/blog/202207/31/20220731170533_48bad.thumb.400_0.jpeg"},
])


# image From URL
message = {
    "role": "user",
    "content": [
        {"type": "text", "text": "Describe the content of this image."},
        {"type": "image", "url": "https://c-ssl.dtstatic.com/uploads/blog/202207/31/20220731170533_48bad.thumb.400_0.jpeg"},
    ]
}

# pdf file From URL
message = {
    "role": "user",
    "content": [
        {"type": "text", "text": "Describe the content of this document."},
        {"type": "file", "url": "https://example.com/path/to/document.pdf"},
    ]
}

# image From base64 data
message = {
    "role": "user",
    "content": [
        {"type": "text", "text": "Describe the content of this image."},
        {
            "type": "image",
            "base64": "AAAAIGZ0eXBtcDQyAAAAAGlzb21tcDQyAAACAGlzb2...",
            "mime_type": "image/jpeg",
        },
    ]
}
```
