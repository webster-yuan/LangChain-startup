## 什么时候该用 Deep Agent？什么时候该用纯 LangGraph？


| 场景 / 需求                                               | 推荐使用                   | 理由                               |
| --------------------------------------------------------- | -------------------------- | ---------------------------------- |
| **简单任务**（查天气、简单计算、单轮问答）                | **纯 LangGraph**           | 更轻量、更可控、模型要求低         |
| **需要严格控制流程**（固定步骤、复杂路由、特定循环逻辑）  | **纯 LangGraph**           | 你可以完全自定义节点和边，流程稳定 |
| **需要 Human-in-the-Loop（人工审核）**                    | **纯 LangGraph**           | 更容易精确控制中断位置和恢复逻辑   |
| **多 Agent 团队协作**（Supervisor + 多个 Worker）         | **纯 LangGraph**（推荐）   | 目前更成熟、可调试性强             |
| **快速原型，任务比较复杂**（需要规划、分解任务）          | **Deep Agent**（尝试）     | 内置规划工具，能快速看到效果       |
| **需要长期记忆 + 文件系统管理**（生成报告、保存中间结果） | **Deep Agent**（理想情况） | 内置虚拟文件系统很方便             |
| **希望 Agent 自动组建子团队**（动态创建子 Agent）         | **Deep Agent**             | 这是它最突出的特性                 |
| **本地小模型（7B~13B）**                                  | **纯 LangGraph**           | Deep Agent 对模型要求高，容易失效  |
| **强模型（GPT-4o、Claude 3.5、Qwen2.5-72B 等）**          | **Deep Agent** 可优先尝试  | 能较好发挥其高级特性               |
| **生产环境 / 需要高稳定性**                               | **纯 LangGraph**           | 更可控、可观测、可调试             |

### 简单实用的判断口诀

**用纯 LangGraph 的情况（推荐优先考虑）：**

- 你想自己掌控流程
- 使用本地中小模型
- 需要高稳定性、生产级可靠性
- 已经对 LangGraph 比较熟悉

**用 Deep Agent 的情况：**

- 你有**较强**的模型（至少 32B+ 或云端顶级模型）
- 想**快速**实现规划、多步骤任务、文件管理、子 Agent 协作
- **愿意接受一定的“黑箱”行为（框架帮你做了很多默认决策）**
- 做研究、原型验证、或非核心功能
